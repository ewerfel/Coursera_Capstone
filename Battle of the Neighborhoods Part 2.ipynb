```python
central_nj__latitude = 40.6962171
central_nj_longitude = -74.2847485
kclusters = 6

# create map
map_clusters = folium.Map(location=[central_nj__latitude, central_nj_longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster , label in zip(nj_merged_with_frequency['Latitude'], nj_merged_with_frequency['Longitude'], nj_merged_with_frequency['Neighborhood'], nj_merged_with_frequency['Cluster Labels'], nj_merged_with_frequency['Descriptive Labels']):
    #if(poi == "Elizabethport, NJ 07206"):
    if True:
        #poi = "E"
        #print(poi + " " + str(lat) + " " + str(lon) + " " + str(cluster))
        cluster = int(cluster)
        label = folium.Popup(str(poi) + "\n" + label, parse_html=True)
        folium.CircleMarker(
            [lat, lon],
            radius=5,
            popup=label,
            color=rainbow[cluster],
            fill=True,
            fill_color=rainbow[cluster],
            fill_opacity=0.7).add_to(map_clusters)


styleList = list([" &nbsp; {} &nbsp; <i class=\"fa fa-map-pin fa-2x\" style=\"color:{}\"></i><br>".format(label_dict.get(counter),value) for counter, value in enumerate(rainbow[0:5])])
#styleList = list([" &nbsp; Cluster {} &nbsp; <i class=\"fa fa-map-marker fa-2x\" style=\"color:{}\"></i><br>".format(counter,value) for counter, value in enumerate(rainbow[:5])])
styleString = "\n<br>".join(styleList)
legend_html =   '''
                <div style="position: fixed; 
                            bottom: 50px; left: 50px; width: 250px; height: 250px; 
                            border:2px solid grey; z-index:9999; 
                            font-size:11px;
                            font-weight:bold;
                            background-color: white;
                            opacity:0.7;
                            "><p font-size:13px>&nbsp; Cluster Legend</p><br>
                            {}                            
                </div>
                '''.format(styleString)     
#legend_html  

map_clusters.get_root().html.add_child(folium.Element(legend_html))
map_clusters

```

```python

tmp = nj_geo_json

# remove ZIP codes not in our dataset
geozips = []
for i in range(len(tmp['features'])):

    if tmp['features'][i]['properties']['GEOID10'] in list(nj_merged_with_frequency['Zip'].unique()):
        geozips.append(tmp['features'][i])


# creating new JSON object
new_json = dict.fromkeys(['type','features'])
new_json['type'] = 'FeatureCollection'
new_json['features'] = geozips

```

```python
central_nj__latitude = 40.6962171
central_nj_longitude = -74.2847485

# create map
map_c = folium.Map(location=[central_nj__latitude, central_nj_longitude], zoom_start=11)
map_c.choropleth(
      geo_data = new_json,
      fill_opacity = 0.7,
      line_opacity=0.2,
      data = nj_merged_with_frequency,
      key_on = 'feature.properties.GEOID10',
      columns = ['Zip','Lawyer Frequency'],
      fill_color='RdYlGn',
      legend_name='Law Office Frequency',
)
folium.LayerControl().add_to(map_c)
# m2.save(outfile='total_establishments_map.html')
map_c
```
